```python
import pandas as pd
import numpy as np

def log_transform(data, columns):
    """ Apply log transformation to specified columns, handling zero values. """
    for col in columns:
        offset = 1e-5  # Small offset to avoid log(0)
        data[col + '_log'] = np.log(data[col] + offset)
    return data


def cap_extreme_values(data, columns):
    """ Cap extreme values for each column at the 99th and 1st percentiles to reduce the influence of outliers. """
    for col in columns:
        upper_limit = data[col].quantile(0.99)  # Cap at the 99th percentile
        lower_limit = data[col].quantile(0.01)  # Cap at the 1st percentile
        data[col] = np.clip(data[col], lower_limit, upper_limit)
    return data

def calculate_z_scores(data, columns):
    """ Calculate Z-scores for specified columns in a DataFrame. """
    z_scores_df = data.copy()
    for col in columns:
        z_scores_df[col + '_z'] = (z_scores_df[col] - z_scores_df[col].mean()) / z_scores_df[col].std()
    return z_scores_df

def compute_role_score(data, columns, weights):
    """ Compute the weighted score for the 'Progressor' role, scaling the result to 0-100. """
    z_score_columns = [col + '_z' for col in columns]
    
    # Calculate weighted sum of Z-scores
    data['progressor_score_raw'] = data[z_score_columns].dot(weights)
    
    # Scale scores to 0-100
    min_score = data['progressor_score_raw'].min()
    max_score = data['progressor_score_raw'].max()
    data['progressor_score'] = ((data['progressor_score_raw'] - min_score) / (max_score - min_score)) * 100
    
    return data

def prepare_final_df(original_df, score_df, role_column):
    """ Prepare the final DataFrame to include player, team, and role score. """
    final_df = original_df[['player', 'team', 'league']].copy()
    final_df[role_column] = score_df[role_column]
    return final_df

# Define the columns and their corresponding weights
data_columns = ['progressive_distance_per_carry', 'touches_def_mid_third_per_team', 'progressive_distance_per_pass', 'percent_total_passes_per_team', 'Total_Cmp%']
weights = np.array([0.25, 0.20, 0.25, 0.15, 0.15])

# Assuming cb_team_stats_df is preloaded with your data
# Cap extreme values before any transformations
cb_team_stats_df = cap_extreme_values(cb_team_stats_df, data_columns)

# Calculate Z-scores on a copy of the original DataFrame
scores_df = calculate_z_scores(cb_team_stats_df, data_columns)

# Compute 'Progressor' role score with new weights
scores_df = compute_role_score(scores_df, data_columns, weights)

# Prepare the final DataFrame for output
cb_scores_df = prepare_final_df(cb_team_stats_df, scores_df, 'progressor_score')

# Print the updated DataFrame to check the results
print(cb_scores_df)
```

```python
import pandas as pd
import numpy as np

def log_transform(data, columns):
    """ Apply log transformation to specified columns, handling zero values. """
    for col in columns:
        offset = 1e-5  # Small offset to avoid log(0)
        data[col + '_log'] = np.log(data[col] + offset)
    return data

def cap_extreme_values(data, columns):
    """ Cap extreme values for each column at the 99th and 1st percentiles to reduce the influence of outliers. """
    for col in columns:
        upper_limit = data[col].quantile(0.99)  # Cap at the 99th percentile
        lower_limit = data[col].quantile(0.01)  # Cap at the 1st percentile
        data[col] = np.clip(data[col], lower_limit, upper_limit)
    return data

def calculate_z_scores_aggressive(data, columns):
    """ Calculate Z-scores for specified defensive columns in a DataFrame. """
    z_scores_df = data.copy()
    for col in columns:
        z_scores_df[col + '_z'] = (z_scores_df[col] - z_scores_df[col].mean()) / z_scores_df[col].std()
    return z_scores_df

def compute_aggressive_defender_score(data, columns, weights):
    """ Compute the weighted score for the 'Aggressive Defender' role, scaling the result to 0-100. """
    z_score_columns = [col + '_z' for col in columns]
    data['aggressive_defender_score_raw'] = data[z_score_columns].dot(weights)
    min_score = data['aggressive_defender_score_raw'].min()
    max_score = data['aggressive_defender_score_raw'].max()
    data['aggressive_defender_score'] = 100 * (data['aggressive_defender_score_raw'] - min_score) / (max_score - min_score)
    return data

# Ensure metrics are calculated before Z-score calculations
def ensure_metrics(data):
    if 'Clr' in data.columns and 'Touches_Mid 3rd_allowed_poss' in data.columns:
        data['blocked_passes_per_100_opp_touch_in_mid_3rd'] = (data['Blocks_Pass'] / data['Touches_Mid 3rd_allowed_poss']) * 100
        data['blocks_per_100_opp_touch_in_mid_3rd'] = (data['Blocks_Sh'] / data['Touches_Mid 3rd_allowed_poss']) * 100
        data['tkl_and_int_per_100_opp_touch_in_mid_3rd'] = (data['Tkl+Int'] / data['Touches_Mid 3rd_allowed_poss']) * 100
    if 'Aerial Duels_Won' in data.columns and 'Aerial Duels_Lost' in data.columns and '90s' in data.columns:
        data['aerial_duels_per_90'] = (data['Aerial Duels_Won'] + data['Aerial Duels_Lost']) / data['90s']
    return data

# Define columns and weights
aggressive_columns = ['blocked_passes_per_100_opp_touch_in_mid_3rd', 'blocks_per_100_opp_touch_in_mid_3rd', 'tkl_and_int_per_100_opp_touch_in_mid_3rd', 'aerial_duels_per_90', 'Aerial Duels_Won%']
weights_aggressive = np.array([0.4, 0.5, 0.4, 0.10, 0.10])

# Assuming cb_team_stats_df is preloaded with your data
cb_team_stats_df = ensure_metrics(cb_team_stats_df)  # Calculate metrics

# Cap extreme values before any transformations
cb_team_stats_df = cap_extreme_values(cb_team_stats_df, aggressive_columns)

# Calculate Z-scores for the new defensive stats
scores_df_aggressive = calculate_z_scores_aggressive(cb_team_stats_df, aggressive_columns)

# Compute 'Aggressive Defender' role score with specified weights
scores_df_aggressive = compute_aggressive_defender_score(scores_df_aggressive, aggressive_columns, weights_aggressive)

# Prepare for merge
scores_df_aggressive['player'] = cb_team_stats_df['player']
scores_df_aggressive['team'] = cb_team_stats_df['team']

# Merge the score into cb_scores_df
cb_scores_df = cb_scores_df.merge(scores_df_aggressive[['player', 'team', 'aggressive_defender_score']], on=['player', 'team'], how='left')

# Print the updated DataFrame to check the results
print(cb_scores_df.head())
```
