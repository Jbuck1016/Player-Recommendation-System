```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial import distance

# Define the URL of the CSV file
url = 'https://raw.githubusercontent.com/Jbuck1016/Player-Recommendation-System/main/Player%20Metrics/Team%20Stats/Team%20Playstyle%20Scores.csv'

# Load the CSV file into a DataFrame directly from GitHub
teams_df = pd.read_csv(url)
teams_df.set_index('team', inplace=True)  # Assuming 'team' is the column with team names

def calculate_team_style_similarity(teams_df):
    scaler = MinMaxScaler()
    score_columns = ['high_press_score', 'deep_block_score', 'narrow_defense_score', 
                     'deep_circulation_score', 'high_retention_score', 'wide_attack_score']
    df_normalized = teams_df.copy()
    df_normalized[score_columns] = scaler.fit_transform(teams_df[score_columns])
    distance_matrix = distance.cdist(df_normalized[score_columns], df_normalized[score_columns], 'euclidean')
    similarity_scores = 100 * (1 - (distance_matrix / distance_matrix.max()))
    # Set index and columns to team names which are now the index of df_normalized
    similarity_df = pd.DataFrame(similarity_scores, index=teams_df.index, columns=teams_df.index)
    return similarity_df

team_style_similarity = calculate_team_style_similarity(teams_df)
team_style_similarity
```

```python
import pandas as pd

def generate_player_shortlist(team, position, role, df_scores, df_team_ratings, df_league_ratings, team_style_similarity):
    role_column = f"{role.lower()}_score"
    candidates = df_scores[(df_scores['team'] != team) & (df_scores[role_column].notna())].sort_values(by=role_column, ascending=False)

    shortlist = []
    team_league = df_team_ratings.loc[df_team_ratings['team'] == team, 'League'].dropna().values[0]

    for _, player in candidates.iterrows():
        player_team = player['team']
        player_league = player['league']
        role_score = player[role_column]
        if isinstance(role_score, pd.Series):
            role_score = role_score.iloc[0]

        team_style_score = team_style_similarity.loc[team, player_team] if (team in team_style_similarity.index and player_team in team_style_similarity.columns) else 0
        if isinstance(team_style_score, pd.Series):
            team_style_score = team_style_score.iloc[0]

        team_quality_score = get_team_quality_similarity(player_team, team, df_team_ratings)
        if isinstance(team_quality_score, pd.Series):
            team_quality_score = team_quality_score.iloc[0]

        league_quality_score = get_league_quality_similarity(player_league, team_league, df_league_ratings)
        if isinstance(league_quality_score, pd.Series):
            league_quality_score = league_quality_score.iloc[0]

        combined_score = (0.5 * role_score +
                          0.2 * team_style_score +
                          0.2 * team_quality_score +
                          0.1 * league_quality_score)

        if isinstance(combined_score, pd.Series):
            combined_score = combined_score.iloc[0]

        shortlist.append({'Player': player['player'], 'Team': player_team, 'Score': combined_score})

    if not shortlist:
        return pd.DataFrame()  # Return an empty DataFrame if no candidates

    df_shortlist = pd.DataFrame(shortlist)
    df_shortlist.sort_values(by='Score', ascending=False, inplace=True)
    return df_shortlist.head(15)






def get_team_quality_similarity(player_team, team, df_team_ratings):
    """
    Calculates the quality similarity score between two teams based on their ratings.

    :param player_team: str - Team of the player being evaluated.
    :param team: str - Team looking to recruit.
    :param df_team_ratings: DataFrame - Contains ratings for each team.
    :return: float - Quality similarity score.
    """
    player_ratings = df_team_ratings[df_team_ratings['team'] == player_team]['rating']
    team_ratings = df_team_ratings[df_team_ratings['team'] == team]['rating']

    if player_ratings.empty or team_ratings.empty:
        print(f"Rating data missing for {player_team} or {team}.")
        return 0

    return 1 - abs(player_ratings.iloc[0] - team_ratings.iloc[0]) / 100

def get_league_quality_similarity(player_league, team_league, df_league_ratings):
    """
    Calculates the quality similarity score between two leagues based on their average ratings.

    :param player_league: str - League of the player being evaluated.
    :param team_league: str - League of the team looking to recruit.
    :param df_league_ratings: DataFrame - Contains average ratings for each league.
    :return: float - League quality similarity score.
    """
    player_league_ratings = df_league_ratings[df_league_ratings['League'] == player_league]['Average_Rating']
    team_league_ratings = df_league_ratings[df_league_ratings['League'] == team_league]['Average_Rating']

    if player_league_ratings.empty or team_league_ratings.empty:
        print(f"League rating data missing for {player_league} or {team_league}.")
        return 0

    return 1 - abs(player_league_ratings.iloc[0] - team_league_ratings.iloc[0]) / 100

# Example usage
# Example usage
team = 'Arsenal'
position = 'FW'
role = 'Poacher'
recommended_players_df = generate_player_shortlist(team, position, role, FW_scores, team_ratings, league_ratings, team_style_similarity)
print("Recommended players:")
recommended_players_df

```
